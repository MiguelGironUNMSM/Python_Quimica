```python
# Lee los datos desde la tabla 2
url = 'https://es.wikipedia.org/wiki/Ley_de_Boyle-Mariotte'
tablas = pd.read_html(url)

data = tablas[2]
# Crea un gráfico de dispersión para mostrar la relación entre P y V
plt.figure(figsize=(6, 4))
plt.scatter(data['V (L)'], data['P (atm)'], marker='o', color='b')
plt.xlabel('Volumen (L)')               #etiqueta eje x
plt.ylabel('Presión (atm)')             #etiqueta eje y
plt.title('Ley de Boyle-Mariotte')      #para el titulo
plt.grid(True) #para las lineas de referencia

# Añade una línea recta para una representación visual de la ley de Boyle
plt.plot(data['V (L)'], data['P (atm)'], color='r', linestyle='--', label='Ley de Boyle-Mariotte')

plt.legend()

plt.show()
# En este código, estamos utilizando un gráfico de dispersión para mostrar cómo cambia la presión (P)
# con respecto al volumen (V). También hemos añadido una línea roja punteada para representar visualmente
# la Ley de Boyle-Mariotte
```
