<b>PROPIEDADES DE LOS LIQUIDOS</b>

image.png

Debido a que su estructura y el comportamiento de sus partículas es diferente; en los gases los particulas poseen fuerzas intermoleculares muy debiles, por lo que se encuentran dispersas; por lo cual son facilmente compresibles; por otra parte los liquidos poseen unas fuerzas intermoleculares mas fuertes las cuales mantienen unidas a las partículas en una estructura ordenada; po lo que hay poco espacio que pueda ser comprimido entre sus partículas.

image.png

La tensión superficial es la tendencia que tienen la superficie de los líquidos a comportarte como una película elástica; esta se genera debido a que las fuerzas intermoleculares que actuan sobre las moléculas en la superficie tiran hacia los costados o al interior; lo cual las acerca dando una apariencia de "tensión"; la tensión superficial disminuye con el incremento de la temperatura; ya que esta aumenta la energía cinética de las moléculas,lo que causa que las fuerzas intermolecualres disminuirían afectando a la tensión superficial;

image.png

Debido a la tensión superficial; las moléculas del agua estan unidas mediante "puentes de hidrógeno", los cuales son enlaces fuertes; por lo que se necesita que el peso sea mayor a esta resistencia para poder hundirse; la navaja posee un peso muy pequeño y ademas tendría que superar la fuerza de tensión superficial de un área extensa en proporción al peso.

image.png

La Cohesión: son las fuerzas de atracción entre moléculas de un mismo líquido; y permiten que permanezcan juntas; por ejemplo las gotas de lluvia, son esfericas; debido a que las moléculas de agua posee una gran cohesión debido a los acción de los puentes de hidrógeno.<br>
La Adhesión: Son las fuerzas de Atracción entre moléculas de diferentes líquidos; responsables de la capacidad de un líquido para expandirse en una superficie. Por ejemplo el mercurio si se derrama en el suelo no se expandira y mojara una gran superficie, sino que se formaran un conjunto de gotitas; esto se debe a que el mercurio posee uan baja adhesión.

image.png

Imaginemos que ponemos azúcar en un vaso con agua; en un inicio las moléculas de agua debido a su interacción con las moléculas de azúcar, debilitan las fuerzas intermoleculares en la superficie del azúcar sólido. Luego debido a la interacción de cargas las moléculas de agua rodean y se incorporan entre las moléculas de azúcar, rodeándolas y separándolas entre sí. Este proceso sigue hasta conseguir una mezcla homogénea.

image.png

Esta frase plantea que  las sustancias con fuerzas intermoleculares similares o compatibles tienen una mayor tendencia a mezclarse y disolverse entre sí; esto se ve claramente entre la diferencia entre solventes polares y apolares; por ejemplo el agua es un gran disolvente polar, por lo que disuelve facilmetne moleculas de azucar; pero no puede disolver aceites ya que estos son apolares.

image.png

Si nos encontramos con un soluto y solvente cuyas fuerzas intramoleculares sean mayores a la fuerza intermolecular entre ellos; la disolución nunca se va a dar; a menos que suministremos de energía en forma de calor al sistema, para que de esta forma las fuerzas intermoleculares superen a las intromoleculares.<br>
En el caso contrario, osea que las fuerzas intermoleculares sean mayores, al efectuarse la disolución los enlaces intramoleculares emitiran energía.
