<b>PROPIEDADES DE LOS LIQUIDOS</b>

image.png

Debido a que su estructura y el comportamiento de sus partculas es diferente; en los gases los particulas poseen fuerzas intermoleculares muy debiles, por lo que se encuentran dispersas; por lo cual son facilmente compresibles; por otra parte los liquidos poseen unas fuerzas intermoleculares mas fuertes las cuales mantienen unidas a las partculas en una estructura ordenada; po lo que hay poco espacio que pueda ser comprimido entre sus partculas.

image.png

La tensin superficial es la tendencia que tienen la superficie de los lquidos a comportarte como una pelcula elstica; esta se genera debido a que las fuerzas intermoleculares que actuan sobre las molculas en la superficie tiran hacia los costados o al interior; lo cual las acerca dando una apariencia de "tensin"; la tensin superficial disminuye con el incremento de la temperatura; ya que esta aumenta la energa cintica de las molculas,lo que causa que las fuerzas intermolecualres disminuiran afectando a la tensin superficial;

image.png

Debido a la tensin superficial; las molculas del agua estan unidas mediante "puentes de hidrgeno", los cuales son enlaces fuertes; por lo que se necesita que el peso sea mayor a esta resistencia para poder hundirse; la navaja posee un peso muy pequeo y ademas tendra que superar la fuerza de tensin superficial de un rea extensa en proporcin al peso.

image.png

La Cohesin: son las fuerzas de atraccin entre molculas de un mismo lquido; y permiten que permanezcan juntas; por ejemplo las gotas de lluvia, son esfericas; debido a que las molculas de agua posee una gran cohesin debido a los accin de los puentes de hidrgeno.<br>
La Adhesin: Son las fuerzas de Atraccin entre molculas de diferentes lquidos; responsables de la capacidad de un lquido para expandirse en una superficie. Por ejemplo el mercurio si se derrama en el suelo no se expandira y mojara una gran superficie, sino que se formaran un conjunto de gotitas; esto se debe a que el mercurio posee uan baja adhesin.
